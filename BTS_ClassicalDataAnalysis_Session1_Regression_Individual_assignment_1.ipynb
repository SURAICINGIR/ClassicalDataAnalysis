# Linear Regression in Statsmodels

## Load the iris dataset

```python
import statsmodels.api as sm
from sklearn import datasets 
```

```python
iris = datasets.load_iris()
print(iris.data.shape)
print(iris.DESCR)
```

```python
iris.feature_names
```

```python
iris
```

```python
import numpy as np
import pandas as pd
```

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names) 
```

```python
iris_df
```

### Regression model with Statsmodels and without a constant:

```python
import statsmodels.api as sm

X = iris_df["sepal width (cm)"]
y = iris_df["sepal length (cm)"]
```

```python
iris_model = sm.OLS(y, X).fit()
```

```python
predictions = iris_model.predict(X) 
```

```python
iris_model.summary()
```

### Interpreting the Table 

there is a 95% confidence; "sepal width (cm)" variable increases by 1, the predicted value of "sepal length (cm)" increases by 1.8717. R-Square Means:Using "sepal width (cm)" variables, the percentage of explain model,0.957

### Regression model with Statsmodels and with a constant:

```python
import statsmodels.api as sm 
```

```python
X = iris_df["sepal width (cm)"]
y = iris_df["sepal length (cm)"]
```

```python
X = sm.add_constant(X) 
```

```python
iris_modelwh = sm.OLS(y, X).fit()
```

```python
predictions = iris_modelwh.predict(X)
```

```python
iris_modelwh.summary()
```

### Interpreting the Table 

there is a 95% confidence; 
"sepal width (cm)" variable increases by 1, the predicted value of "sepal length (cm)" decrease by 0.2089 and y-intercept 6.4812
R-Square Means:Using "sepal width (cm)" variables, the percentage of explain model,0.012
