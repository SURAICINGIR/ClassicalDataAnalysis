# 1. Data

Let's take a look again at some data - Advertising - , ask some questions about that data, and then use linear regression (with one and multiple variables) to answer those questions!

```python
# imports
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read data into a DataFrame
df = pd.read_csv('http://www-bcf.usc.edu/~gareth/ISL/Advertising.csv', index_col=0)
df.head()
```

```python
df
```

# 2 - Linear Regression with multiple variables in Statsmodels

## 2.1. Regression model with multiple variables and with a constant for Situation 1

```python
Y=df['sales']
X=df[['radio','newspaper']]
```

```python
X=sm.add_constant(X)
```

```python
model = sm.OLS(Y, X).fit()
```

```python
prediction=model.predict(X)
prediction
```

```python
model.summary()
```

```python
model.params
```

## 2.2. Regression model with multiple variables and with a constant for Situation 2

> Put your code here

```python
x1=df[['TV','radio','newspaper']]
y1=df['sales']
```

```python
x1=sm.add_constant(x1)
```

```python
model1 = sm.OLS(y1, x1).fit()
```

```python
prediction1=model1.predict(x1)
prediction1
```

```python
model1.summary()
```

```python
model1.params
```

## 2.3. Regression model with multiple variables and without a constant for Situation 1

```python
Y1=df['sales']
X1=df[['radio','newspaper']]
```

```python
X1.head()
```

```python
model2=sm.OLS(Y1, X1).fit()
```

```python
model2.summary()
```

```python
model2.params
```

```python
prediction2=model2.predict(X1)
prediction2
```

## 2.4. Regression model with multiple variables and without a constant for Situation 2

```python
X2=df[['TV','radio','newspaper']]
Y2=df['sales']
```

```python
model3=sm.OLS(Y2,X2).fit()
```

```python
model3.summary()
```

```python
model3.params
```

```python
prediction3=model3.predict(X2)
prediction3
```

### Interpreting the fitness of the models

The model have more higher R-squared=0.982 and the model calculated use all of variables (TV,radio,newspaper).
